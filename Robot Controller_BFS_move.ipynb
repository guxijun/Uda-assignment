```python
import random as rd

env_data = [
    [3, 2, 2, 2, 2, 2, 2, 2, 1],
    [0, 0, 2, 2, 2, 2, 2, 0, 0],
    [2, 0, 0, 2, 2, 2, 0, 0, 2],
    [2, 2, 0, 0, 2, 0, 0, 2, 2],
    [2, 2, 2, 0, 0, 0, 2, 2, 2]]

'''
一、算法
1.随机寻址
2.广度优先算法（BFS）单向寻址
3.广度优先算法（BFS）双向寻址

二、数据和结构定义
1.“点”
    命名：loc_...
    数据类型：tuple（int_row, int_col）
        int_row: “点”的行值，范围：[0,len(env_data)-1]
        int_col: “点”的列值，范围：[0,len(env_data[0])-1]
2.“点”的字典
    命名：loc_map_...，routine_...
    数据类型：list of tuple（[loc_, loc_, ...]）
3.“点”的状态矩阵
    命名：env_state
    数据类型：list of str，结构同env_data
    取值：a：未寻址，b：正在寻址点位，c：已寻址
4.函数参数传递规则
    避免：函数参数命名==全局变量名称
    方法：函数参数命名=全局变量名称+下划线_
'''

# 通过env_data初始化env_state
env_state = []
for i in range(len(env_data)):
    row_state = []
    for j in range(len(env_data[i])):
        row_state.append('a')
    env_state.append(row_state)
# print(env_state)

# TODO 6设置起点、目标点的字典
loc_map = {}
loc_map = {'start': (0, 8), 'destination': (0, 0)}
# TODO 7保存机器人当前的位置
robot_start = None
robot_start = loc_map['start']
robot_destination = loc_map['destination']


# 定义判断“是否超出迷宫边界”函数
def is_outof_boundary(env_data_, int_row, int_col):
    outof_boundary = (int_row < 0 or int_row > len(env_data_) - 1) or (int_col < 0 or int_col > len(env_data_[0]) - 1)
    return outof_boundary


# TODO 9-1：定义判断“给定方向移动可行性”的函数（用于随机寻址）
def is_move_valid_random(env_data_, loc, act):
    """
    Judge wether the robot can take action act
    at location loc.

    Keyword arguments:
    env -- list, the environment data
    loc -- tuple, robots current location
    act -- string, robots meant action
    """

    loc_row, loc_col = loc
    des_row = None
    des_col = None
    move_valid = None
    # print(loc_row)
    # print(loc_col)

    # 根据方向，计算目标点坐标
    if act == 'u':
        des_row = loc_row - 1
        des_col = loc_col
    elif act == 'd':
        des_row = loc_row + 1
        des_col = loc_col
    elif act == 'l':
        des_row = loc_row
        des_col = loc_col - 1
    elif act == 'r':
        des_row = loc_row
        des_col = loc_col + 1
    else:
        print('invalid action')

    # 根据目标点坐标和值，设置is_move_valid变量值
    if is_outof_boundary(env_data_, des_row, des_col) or env_data_[des_row][des_col] == 2:
        move_valid = False
    else:
        move_valid = True

    return move_valid


# TODO 9-2：定义判断“给定方向移动可行性”的函数（用于BFS算法寻址）
def is_move_valid_bfs(env_data_, loc, act):
    """
    Judge wether the robot can take action act
    at location loc.

    Keyword arguments:
    env -- list, the environment data
    loc -- tuple, robots current location
    act -- string, robots meant action
    """

    loc_row, loc_col = loc
    des_row = None
    des_col = None
    move_valid = None
    # print(loc_row)
    # print(loc_col)

    # 根据方向，计算目标点坐标
    if act == 'u':
        des_row = loc_row - 1
        des_col = loc_col
    elif act == 'd':
        des_row = loc_row + 1
        des_col = loc_col
    elif act == 'l':
        des_row = loc_row
        des_col = loc_col - 1
    elif act == 'r':
        des_row = loc_row
        des_col = loc_col + 1
    else:
        print('invalid action')

    # 根据目标点坐标和值，设置is_move_valid变量值
    if is_outof_boundary(env_data_, des_row, des_col) or env_data_[des_row][des_col] == 2 \
            or env_state[des_row][des_col] == 'b' or env_state[des_row][des_col] == 'c':
        move_valid = False
    else:
        move_valid = True

    return move_valid


## TODO 10-1：定义生成“可行移动集合”的函数（用于随机寻址）
def valid_actions_random(env_data_, loc):
    valid_actions_list = []
    if is_move_valid_random(env_data_, loc, 'u'):
        valid_actions_list.append('u')
    if is_move_valid_random(env_data_, loc, 'd'):
        valid_actions_list.append('d')
    if is_move_valid_random(env_data_, loc, 'l'):
        valid_actions_list.append('l')
    if is_move_valid_random(env_data_, loc, 'r'):
        valid_actions_list.append('r')

    return valid_actions_list


## TODO 10-2：定义生成“可行移动集合”的函数（用于BFS算法寻址）
def valid_actions_bfs(env_data_, loc):
    valid_actions_list = []
    if is_move_valid_bfs(env_data_, loc, 'u'):
        valid_actions_list.append('u')
    if is_move_valid_bfs(env_data_, loc, 'd'):
        valid_actions_list.append('d')
    if is_move_valid_bfs(env_data_, loc, 'l'):
        valid_actions_list.append('l')
    if is_move_valid_bfs(env_data_, loc, 'r'):
        valid_actions_list.append('r')

    return valid_actions_list


##TODO 11定义“移动机器人”的函数
def move_robot(loc, act):
    loc_row, loc_col = loc
    des_row = None
    des_col = None
    # print(loc_row)
    # print(loc_col)
    # 根据方向，计算目标点坐标
    if act == 'u':
        des_row = loc_row - 1
        des_col = loc_col
    elif act == 'd':
        des_row = loc_row + 1
        des_col = loc_col
    elif act == 'l':
        des_row = loc_row
        des_col = loc_col - 1
    elif act == 'r':
        des_row = loc_row
        des_col = loc_col + 1
    else:
        print('invalid action')

    # 移动后，将出发点对应state修改为c，到达点对应state修改为b
    # 由于调用move_robot函数前，会预先判断移动可行性，所以执行move_robot函数后，不会出现out of boundary情况
    env_state[loc_row][loc_col] = 'c'
    env_state[des_row][des_col] = 'b'
    # return到达点
    loc_des = (des_row, des_col)
    return loc_des


##TODO 12-1：随机寻址
def choose_actions_random(env_data_, loc):
    robot_loc = loc
    routine = [robot_loc]
    for i in range(300):
        valid_actions_list = valid_actions_random(env_data_, robot_loc)
        robot_loc = move_robot(robot_loc, rd.choice(valid_actions_list))
        routine.append(robot_loc)
        if env_data_[robot_loc[0]][robot_loc[1]] == 3:
            print('在第{}回合找到宝藏'.format(i + 1))
            break
    print(routine)


# 运行：随机寻址
# choose_actions_random(env_data, robot_start)


##TODO 12-2：算法一：广度优先算法（BFS）单向寻址
def choose_actions_bfs_single(env_data_, loc):
    start_loc = loc
    routine = [start_loc]
    routine_index = 0
    # 启动寻址，设置起点状态为b
    env_state[start_loc[0]][start_loc[1]] = 'b'

    # 循环寻址，直到出现loc_map['destination']
    while routine.count(robot_destination) == 0:
        temp_routine = []
        for i in range(len(routine) - routine_index):
            loc_state_b = routine[len(routine) - 1 - i]
            valid_actions_list = valid_actions_bfs(env_data_, loc_state_b)
            if valid_actions_list == []:
                continue
            for action in valid_actions_list:
                robot_loc = move_robot(loc_state_b, action)
                temp_routine.append(robot_loc)
        # 将temp_routine追加到routine，注意：不能直接append！！！
        for loc_in_temp_routine in temp_routine:
            routine.append(loc_in_temp_routine)
        # 更新routine_index
        routine_index = len(routine) - len(temp_routine)

    print(routine)


# 运行：广度优先算法（BFS）单向寻址
choose_actions_bfs_single(env_data, robot_start)
```
